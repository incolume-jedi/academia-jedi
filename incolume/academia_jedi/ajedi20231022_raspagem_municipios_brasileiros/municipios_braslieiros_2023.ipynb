```python
import pandas as pd
from enum import Enum
import datetime as dt
from dataclasses import dataclass
from pathlib import Path
```

```python
# diretório para armazenamento e leitura de conteúdos gerados.

data_path = Path().resolve().parents[2].joinpath('data_files')
display(data_path.is_dir())
data_path
```

```python
# definição de regiões

Regiao = Enum('Regiao', 'NORTE NORDESTE CENTROESTE SUDESTE SUL'.split())

Regiao
```

```python
# definição de Estado


@dataclass
class Estado:
    sigla: str
    estado: str
    capital: str
    regiao: Regiao
```

```python
# Definição de município


@dataclass
class Municipio:
    estado: Estado = ''
    nome: str = ''
    data: dt.datetime = None
    mes: int = 0
    dia: int = 0
```

```python
# carregamento de dados dos estados

estados = pd.read_json(data_path / 'json' / 'estados_br.json')
estados
```
