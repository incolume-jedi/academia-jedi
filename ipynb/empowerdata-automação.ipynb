## Gerar analises
- cotação máxima
- cotação mínima
- cotação atual

````python
def ticker_values(ticker: str):
    dados = yfinance.Ticker(ticker)
    tabela = dados.history('6mo')
    fechamento = tabela.Close
    fechamento.plot()
    return ticker, fechamento.min(), fechamento.max(), fechamento[-1]
````

````python
ticker_values('ITSA4.SA')
````

````python
try:
    import pyautogui
    import pyperclip
except (ImportError, ModuleNotFoundError):
    !pip install pyautogui
    !pip install pyperclip

    import pyautogui
    import pyperclip
````

````python
pyautogui.PAUSE = 2

pyautogui.hotkey('ctrl', 't')
pyperclip.copy('www.gmail.com')
pyautogui.hotkey('ctrl', 'v')
pyautogui.hotkey('enter')

# pyautogui.position()
# 
pyautogui.click(x=2021, y=206)
pyperclip.copy('gestor@gmail.com')
pyautogui.hotkey('ctrl', 'v')
pyautogui.hotkey('tab')

pyperclip.copy('Assunto: Análises diárias')
pyautogui.hotkey('ctrl', 'v')
pyautogui.hotkey('tab')


message = """Prezado gestor,

Segue em anexo a análise dos últimos 6 meses da ação {}:

- Cotação mínima: R$ {:2.f}
- Cotação máxima: R$ {:2.f}
- Cotação atual: R$ {:2.f}

Em caso de dúvidas estou a disponsição.

Att.
Brito
""".format('JBSS3.SA', ticker_values('JBSS3.SA'))

pyperclip.copy(message)
pyautogui.hotkey('ctrl', 'v')
pyautogui.hotkey('ctrl', 'enter')
````
